## Analisando dados de obras realizadas no município de Contagem
A ideis é correlacionar os anos de uma gestão e o volume de obras realizadas pela prefeitura.

fonte:https://dadosabertos.tce.mg.gov.br/view/xhtml/paginas/telaInicio.xhtml

```python
import pandas as pd
import seaborn as sns

url_2017 = 'https://raw.githubusercontent.com/Gabriel-datapy/Obras_Contagem/master/2017.3118601.despesa.despesa.csv'
url_2018 = 'https://raw.githubusercontent.com/Gabriel-datapy/Obras_Contagem/master/2018.3118601.despesa.despesa.csv'
url_2019 = 'https://raw.githubusercontent.com/Gabriel-datapy/Obras_Contagem/master/2019.3118601.despesa.despesa.csv'
url_2020 = 'https://raw.githubusercontent.com/Gabriel-datapy/Obras_Contagem/master/2020.3118601.despesa.despesa.csv'
```

```python
df_2017 = pd.read_csv(url_2017,sep=';',usecols=['num_anoexercicio','dsc_naturezadespesa','vlr_previsto','vlr_acrescimo'])
df_2018 = pd.read_csv(url_2018,sep=';',usecols=['num_anoexercicio','dsc_naturezadespesa','vlr_previsto','vlr_acrescimo'])
df_2019 = pd.read_csv(url_2019,sep=';',usecols=['num_anoexercicio','dsc_naturezadespesa','vlr_previsto','vlr_acrescimo'])
df_2020 = pd.read_csv(url_2020,sep=';',usecols=['num_anoexercicio','dsc_naturezadespesa','vlr_previsto','vlr_acrescimo'])

df_juncao = pd.concat([df_2017,df_2018,df_2019,df_2020])
```

```python
df_juncao.columns = ['Ano de Exercício','Natureza da Despesa','Valor Previsto','Valor Acrescido']
df_juncao.head()
```

```python
df_juncao['Natureza da Despesa'].unique()
```

```python
df_obras = df_juncao[df_juncao['Natureza da Despesa'].str.contains('OBRAS')]


df_obras.head()
```

```python
colunas_obras = df_obras['Natureza da Despesa'].count()
n = 1

#while n<(colunas_obras-1):
valor = df_obras.iloc[n,1][14:]
df_obras.iloc[n,1] = valor

df_obras.head()
```
